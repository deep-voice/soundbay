```python
from glob import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_to_files = "/Users/shai/personal/deepvoice/Fish_Rach/Results/res_jul10/Inference*.csv"
df = pd.DataFrame()
for f in glob(path_to_files):
    tmp_df = pd.read_csv(f)
    tmp_df['file'] = os.path.basename(f).split('.')[0].split('_')[-1]
    df = pd.concat([df, tmp_df], axis=0)
df.head()
```

```python
# Group by 'original_file' and create a histogram for each group
plt.figure(figsize=(10, 6))  # Adjust size as necessary
df.groupby('file')['1'].plot(kind='hist', alpha=0.5, bins=10, legend=True)

# Customize the plot
plt.title('Histogram of Percentage Values by Original File')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend(title='Original File')

# Show plot
plt.show()
```
